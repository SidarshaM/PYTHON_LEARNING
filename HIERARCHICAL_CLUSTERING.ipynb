```python
import pandas as pd
import numpy as np

data=pd.read_csv("/content/clustering - Sheet1 (1).csv")
```

```python
data
```

```python
# deleteing unwanted un necessary columns
data
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
data_features=data.drop("Country",axis=1)
data_Scaled = sc.fit_transform(data_features)
```

```python
#missing,outliers,null  chcek
```

```python
data_Scaled
```

```python
from sklearn.cluster import AgglomerativeClustering
ac=AgglomerativeClustering(n_clusters=2,linkage="ward")
data_AC=ac.fit_predict(data_Scaled)
data_AC
```

```python
data["CLUSTERS"]=data_AC
data
```

```python
import seaborn as sns
sns.scatterplot(x="GDP per Capita",y="Literacy Rate",data=data,hue="CLUSTERS")
```

```python
from sklearn.metrics import silhouette_score

# Compute silhouette score
sil_score = silhouette_score(data_Scaled, data['CLUSTERS'])
sil_score
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt
```

```python
# Perform hierarchical clustering
Z = linkage(data_Scaled, method='ward')

# Plot the dendrogram
plt.figure(figsize=(12, 8))
dendrogram(Z, labels=data['Country'].tolist(), leaf_rotation=90)
plt.title('Dendrogram of Agglomerative Hierarchical Clustering')
plt.xlabel('Countries')
plt.ylabel('Distance')
plt.show()
```
