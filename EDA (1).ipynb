```python
#1.Which country has the most number of players (score :1)
fifa_data['Nationality'].value_counts()# value_counts returns count of unique values
```

```python
# Number of the players in the top 5 countries having most number of players.
list(fifa_data['Nationality'].value_counts()[0:5].keys())
```

```python
#number of players in the top 5 countries
list(fifa_data['Nationality'].value_counts()[0:5])
```

```python
#2.Plot a bar chart of 5 top countries with the most number of players. (score :1)
x=list(fifa_data['Nationality'].value_counts()[0:5].keys())
y=list(fifa_data['Nationality'].value_counts()[0:5])
plt.figure(figsize=(8,4))
plt.bar(x,y,color=['orange','yellow','BLUE','VIOLET','pink','red'],ec='red',width=0.9)
plt.title("COUNTRY VS NUMBER OF PLAYERS",fontsize=15,color="green")
plt.xlabel('Country Names ',fontsize=10)
plt.ylabel('Number of Players',fontsize=10)
plt.show()
```

```python
fifa_data['Wage']=fifa_data['Wage'].replace('€','').replace('K','e3').replace('M','e6')#converting to numeric
fifa_data['Wage'].head()
```

```python
#3.Which player has the highest salary? (score :1)
# create a subset with name and salary details
salary_players=fifa_data[['Name','Wage']]
salary_players
```

```python
salary_players=salary_players.sort_values(by="Wage",ascending=False)
salary_players.head()#return first 5 rows from  sorted array
```

```python
#player with the highest salary is
list(salary_players["Name"].head(1)) # return only the top row value .
```

```python
#4.Plot a histogram to get the salary range of the players. (score :1)


plt.figure(figsize=(8, 5))
plt.hist(salary_players['Wage'].dropna()[0:200], bins=5, color='orange', edgecolor='black')#top 5 
plt.title('Salary Range of Players', fontsize=16)
plt.xlabel('number of players', fontsize=12)
plt.ylabel('Number of Players', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

```python

plt.figure(figsize=(8, 5))
sns.barplot(x=list(salary_players['Name'][0:5]),y=list(salary_players['Wage'][0:5]),ec='black',lw=1)
plt.title('Salary Range of Players', fontsize=16)
plt.xlabel('Player Name', fontsize=12)
plt.ylabel('Salary', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

```python
#5.Who is the tallest player in the fifa? (score :1)
player_height=fifa_data[['Name','Height']].dropna()#remove NaN
player_height=player_height.sort_values(by="Height",ascending=False)
#player_height.head(5)
player_height.head(1)
```

```python
plt.figure(figsize=(8, 6))
x = fifa_data['Preferred Foot'].value_counts().index
y = fifa_data['Preferred Foot'].value_counts().values

# Bar chart
sns.barplot(x=x, y=y,color='orange')
plt.title('Preferred Foot Distribution')
plt.ylabel('Number of Players')
plt.xlabel('Preferred Foot')
plt.show()
```
